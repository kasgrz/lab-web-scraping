```python
link = book.find("h3").find("a")["href"]
print(link)
```

```python
from urllib.parse import urljoin

book_url = urljoin(
    "https://books.toscrape.com/",
    link
)

print(book_url)
```

```python
response = requests.get(book_url)
detail_soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(detail_soup.title)
```

```python
print(detail_soup.prettify()[:1000])
```

```python
table = detail_soup.find("table")
print(table.prettify())
```

```python
rows = table.find_all("tr")
print(len(rows))
```

```python
print(rows[0])
```

```python
header = rows[0].find("th").get_text()
value = rows[0].find("td").get_text()

print(header)
print(value)
```

```python
breadcrumb = detail_soup.find("ul", class_="breadcrumb")
genre = breadcrumb.find_all("li")[2].get_text(strip=True)

print(genre)
```

```python
table = detail_soup.find("table")

upc = None
availability = None

for row in table.find_all("tr"):
    header = row.find("th").get_text(strip=True)
    value = row.find("td").get_text(strip=True)

    if header == "UPC":
        upc = value

    elif header == "Availability":
        availability = value

print("UPC:", upc)
print("Availability:", availability)
```

```python
description_header = detail_soup.find(id="product_description")
```

```python
print(description_header)
```

```python
description = description_header.find_next_sibling("p").get_text(strip=True)

print(description)
```

```python
book_data = {
    "UPC": upc,
    "Title": title,
    "Price (£)": price,
    "Rating": rating,
    "Genre": genre,
    "Availability": availability,
    "Description": description
}

print(book_data)
```
